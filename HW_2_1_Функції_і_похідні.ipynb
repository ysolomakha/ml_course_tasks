### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
f = 10 * x - 5
derivative_f = sp.diff(f, x)
```

```python
print("Функція: f(x) =", f)
print("Похідна: f'(x) =", derivative_f)
```

```python
x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = 10 * np.ones_like(x)
plt.plot(x, f, label='f(x) = 10x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

За графіком видно що за кожну годину оцінка зростає на 10 - в х=10 ми отримуємо оцінку 95, при тому що в х=0 ми мали оцінку -5.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')

f_sin = sp.sin(x)
f_ln = sp.ln(x)

derivative_f_sin = sp.diff(f_sin, x)
derivative_f_ln = sp.diff(f_ln, x)
```

```python
print("Функція: f(x) =", f_sin)
print("Похідна: f'(x) =", derivative_f_sin)
```

```python
print("Функція: f(x) =", f_ln)
print("Похідна: f'(x) =", derivative_f_ln)
```

```python
x_sin = np.linspace (0, 2 * np.pi, 100)
f_sin = np.sin(x_sin)
df_sin = np.cos(x_sin)

plt.plot(x_sin, f_sin, label='f(x) = sin(x)', color='blue')
plt.plot(x_sin, df_sin, label="f'(x) = cos(x)", color='green')
plt.title("Графік функції sin(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x_ln = np.linspace(0.1, 5, 100)
f_ln = np.log(x_ln)
df_ln = 1 / x_ln
plt.plot(x_ln, f_ln, label='f(x) = ln(x)', color='blue')
plt.plot(x_ln, df_ln, label="f'(x) = 1/x", color='green')
plt.title("Графік функції ln(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f = x**2 - 5*x + 1
derivative_f = sp.diff(f, x)
print("Функція: f(x) =", f)
print("Похідна: f'(x) =", derivative_f)
```

```python
x = np.linspace(0, 100, 100)
f = x**2 - 5*x + 1
df = 2*x - 5

plt.plot(x, f, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x, df, label="f'(x) = 2x - 5", color='green')
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()  
```
